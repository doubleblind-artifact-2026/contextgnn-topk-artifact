```python
import glob, os, shutil
from pathlib import Path

repo_name = "<TODO: name>"
working_repo = Path("/kaggle/working") / repo_name
matches = glob.glob(f"/kaggle/input/notebooks/*/output/{repo_name}")

if matches:
    if working_repo.exists():
        shutil.rmtree(working_repo)
    
    shutil.copytree(matches[0], working_repo)
else:
    os.system(f"git clone -b main --single-branch <TODO: link> {working_repo}")

working_log = Path("/kaggle/working/output.log")
log_matches = glob.glob("/kaggle/input/notebooks/*/output/output.log")

if log_matches:
    shutil.copy2(log_matches[0], working_log)

os.chdir(working_repo)
```
